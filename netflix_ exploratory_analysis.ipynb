### Then, what caught my attention was the years column and  I wanted to count the number of releases for each year to see exactly when the numbers increased, or when the numbers dropped. 

```python
count_release=titles_df.groupby("release_year").agg(number_of_release_each_year=("release_year","count")).reset_index()
count_release
```

### The graph below is quite explanatory, the increase in the number of both movies and shows is steady up until 2015, when there seemed to have a rampant surge, especially for movies. The reason for this might be that during those years the platform itself became extremely popular, and at the same time it started producing its own movies and shows. 

```python
plt.figure(figsize=(20,10))
sns.countplot(data= titles_df, y= 'release_year', hue='type')
plt.ylim(20,65)
plt.title('How many movies/shows are released each year?')
plt.show()
```

### As we can see in the dataset, each movie/show has a rating, so I wanted to see if there is any correlation between each year, the total number of releases per year, the maximum rating and minimum rating. As we can see, for movies/shows released in the second half of the 20th century, because their number is so little, the ratings do not differ much from eachother, whereas as we get closer to present times, since the number of releases has increased exponentially, the ratings also vary from very low to very high. 

```python
titles_df.groupby("release_year").agg(total_of_release_each_year=("release_year","count"),max_rating=("imdb_score","max"),min_rating=("imdb_score","min")).reset_index()
```

### Even when we look at the runtime for each year, we can easily observe how the industry changed and developed each year.

```python
titles_df.groupby("release_year").agg(runtime=("runtime","sum"),max_run_time=("runtime","max"),min_run_time=("runtime","min")).reset_index()
```
